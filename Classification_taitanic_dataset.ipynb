```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
!pip install https://github.com/pandas-profiling/pandas-profiling/archive/master.zip
from pandas_profiling import ProfileReport
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB, MultinomialNB
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/insaid2018/Term-1/master/Data/Casestudy/titanic_train.csv')
data.head()
```

```python
profile=ProfileReport(data)
profile.to_file('taitanic.html')
```

```python
data.sample(frac=1)
```

```python
from sklearn.model_selection import train_test_split
train, test=train_test_split(data, test_size=0.2)
```

```python
train.shape, test.shape
```

```python
profile2=ProfileReport(train)
profile2.to_file('train.html')
```

```python
train.isna().sum()
```

```python
train.Age.fillna(train.Age.mean(), inplace=True)
```

```python
train.Embarked.value_counts()
```

```python
train.Embarked.fillna('S', inplace=True)
```

```python
train=pd.get_dummies(train,columns=['Pclass','Sex', 'Embarked'],drop_first=True)
```

```python
train.head()
```

```python
lr=LogisticRegression()
```

```python
train_cols=['Age','Sex_male', 'SibSp', 'Parch', 'Fare']
```

```python
target_col='Survived'
```

```python
lr.fit(train[train_cols],train[target_col])
```

```python
lr.predict(train[train_cols])
```
